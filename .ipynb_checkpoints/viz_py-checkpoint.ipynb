```python
full = pd.read_csv('AS_all_full2.csv')
del full['X']
del full['Unnamed: 0']
full.head(3)
```

```python
# Where growth type was surveyed, count as value; otherwise, NaN
full['encrusting_real'] = np.where(full['encrusting'] == 0, np.nan, full['encrusting_real'])
full['free_livin_real'] = np.where(full['free_livin'] == 0, np.nan, full['free_livin_real'])
full['massive_real'] = np.where(full['massive'] == 0, np.nan, full['massive_real'])
full['plate_real'] = np.where(full['plate'] == 0, np.nan, full['plate_real'])
full['scler_real'] = np.where(full['scler'] == 0, np.nan, full['scler_real'])
```

```python
full.surveyed.sum()/len(full)
```

```python
full['surveyed_ID'] = np.where(full.surveyed != 0, full.ID, 'Not surveyed')
```

```python
full.to_csv('AS_all_full.csv', index=False)
```
